```python
##cargamos la libreria miceforest

import miceforest as mf
from sklearn.datasets import load_iris  ##extraigo el dataset iris
import numpy as np                      ## importamos numpy, es para calculos
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris
```

```python
## convertir un archivo de tipo sklearn a tipo data frame para que python me deje trabajar en el archivo

df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target
```

```python
df_iris.columns
```

```python
##Renombrar la columna target para species

df_iris.rename({"target":"species"}, inplace=True, axis=1)
df_iris
```

```python
##transformo el tipo de dato del campo species

df_iris['species'] = df_iris['species'].astype('category')
df_iris.info()
```

```python
##asigno valores nulos a la columna de especies, de forma aleatorio, no se a que columnas se le van a asignar
## el state es la profundidad del arbol
## el perc es el 25%

iris_amp = mf.ampute_data(df_iris, perc=0.25,random_state=1991)
iris_amp
```

```python
## crear kernel, va a coger columna por columna y va a rellenar los datos nulos con la correlacion que encuentre entre los vlares faltantes de la primera columna con el valor faltante de la segunda columna, para eso sirve el arbol

kds = mf.ImputationKernel(
    iris_amp,                      ##Nombre de dataframe para entrenar el modelo
    random_state=1991              ##Valores aleatorios para entrenar modelo

)

##correr el algoritmo MICE durante 2 interacciones

kds.mice(2)

##retorna el dataset completo

iris_completo = kds.complete_data()
iris_completo
```
